```python
# Using train-test split module

import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

from sklearn.linear_model import LogisticRegression
import csv 


SEEDS_DATA_URL ='https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt'

seeds_df = pd.read_csv(SEEDS_DATA_URL,sep="\s+",names = ["Area","Perimeter", 
                                                   "Compactness", 
                                                   "Length","Width",  
                                                   "Asymmetry Coefft",
                                                   "Kernel Groove Length",
                                                   "Target"])
y = seeds_df.Target

# create training and testing dataset
X_train, X_test, y_train, y_test =train_test_split(seeds_df, y, test_size=0.9, random_state = 19)

print ("The shape of X_train & y_train are : ",X_train.shape, y_train.shape)
print ("The shape of X_test & y_test are : ",X_test.shape, y_test.shape)

# fit a model
lr = LogisticRegression()
lr.fit(X_train, y_train)
prediction = lr.predict(X_test)  # will output array with integer values.

#display(preds)    
#display("The predicted dataset shape :",prediction.shape)



display(" The actual dataset array :", y_test)
display("The predicted dataset array :",prediction)

# The line / model
#plt.scatter(y_test, preds)
plt.plot(y_test,prediction, 'b*',markersize =15)

print("Model Accuracy : ",(100*lr.score(X_test,y_test)),'%')


np.savetxt('output_1_individual_predictors.csv', prediction.transpose(),
                                                delimiter=',', fmt='%d')
```

```python
#  Trying to learn support vector machine algorithm

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import svm

import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.pyplot import scatter

%matplotlib inline
```

```python
Area_KGLength = seeds_df[['Area','Kernel Groove Length']]. as_matrix ()
type_target = np.where( seeds_df['Target'] == 1,2,3)
```

```python
model = svm.SVC(kernel = 'linear')
model.fit(Area_KGLength, type_target)
```

```python
w = model.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(6, 22)
yy = a * xx - (model.intercept_[0]) / w[1]
```

```python
b = model.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = model.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
```
